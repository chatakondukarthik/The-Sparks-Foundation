Create the Decision Tree classifier and visualize it graphically.

Lets import some necessary files

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

sklearn.datasets is comprised of many datasets like The Olivetti faces dataset, Iris plants dataset, Boston house prices dataset, Forest covertypes and many more

```python
from sklearn import datasets
df=datasets.load_iris()
```

```python
df.data, df.target
```

```python
x = pd.DataFrame(df.data,columns=df.feature_names)
y = df.target
x,y
```

# Training the model

Splitting the dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test=train_test_split(x,y,random_state=1,stratify=y,test_size=0.25)
```

I have dived the dataset in the ratio 75:25, where 75% of data is used for training and the other 25% is used for testing

Fitting the values

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree.fit(X_train,y_train)
```

# Accuracy

```python
y_pred=tree.predict(X_test)
y_pred
```

```python
data = {'Real' : y_test, 'Prediction' : y_pred}
pred = pd.DataFrame(data)
pred
```

```python
print("The accuracy of model is ",tree.score(X_test,y_test))
```

# Plotting the tree

```python
fig=plt.figure(figsize=(25,20))
plot_tree(tree,feature_names=df.feature_names,class_names=df.target_names,filled=True)
```
